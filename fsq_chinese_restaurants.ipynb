```python
import pandas as pd
```

```python
restaurants = pd.read_csv("chinese_restaurants_cleaned.csv")
print(restaurants.shape)
restaurants.head()
```

```python
restaurants.columns
```

```python
# restaurants that are probably still open: 'date_closed' is null

still_open = restaurants[restaurants['date_closed'].isna()]
print(f'open restaurants: {len(still_open)}')
```

```python
# how many of the open restaurants have a website listed?

has_website = still_open[still_open['website'].notnull()]
print(has_website.shape[0])
has_website
```

```python
has_website.to_csv("website_restaurants.csv", index=False)
```

```python
us_restaurants = restaurants[restaurants['country'] == 'US']
print(f'US restaurants: {len(us_restaurants)}')
```

```python
still_open_us = us_restaurants[us_restaurants['date_closed'].isna()]
print(f'open US restaurants: {len(still_open_us)}')
```
